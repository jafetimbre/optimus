```
dim_series = 100

t = [ i for i in range(dim_series) ] 

s1 = np.random.normal(-1, 2, dim_series)
s2 = np.random.normal(0, 0.5, dim_series)
s3 = s1 + s2
s4 = 3 * s1 + s2
```

```
data = pd.DataFrame(data={
    't': t,
    'a': s1,
    'b': s2,
    'c': s3,    # a + b
    'd': s4  # 3a+b
})
data
```

```
fig, ax = plt.subplots(figsize=(8,5))

a_mean = np.mean(data.a)
a_std = np.std(data.a)

ax.plot(data.a)
ax.plot([a_mean]*len(data.a), linestyle='--')
ax.fill_between(data.t, a_mean-a_std, a_mean+a_std, alpha=0.2, color='gray')
plt.show()
```

```
fig, ax = plt.subplots(2, figsize=(8,10))

ax[0].hist(data.c)
ax[1].hist(data.d)
```

```
fig, ax = plt.subplots(figsize=(8,5))

data.c.plot.density()
data.d.plot.density()
```

```
fields = ['realgdp', 'realcons', 'realinv', 'realgovt']

data = pd.read_csv('https://raw.githubusercontent.com/jafetimbre/optimus/master/data/macrodata.csv', usecols=fields)
data.dropna(inplace=True)
data
```

```
norm_data = (data-data.min())/(data.max()-data.min())
norm_data
```

```
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(norm_data.realgdp, color='red')
ax.plot(norm_data.realcons, color='green')
ax.plot(norm_data.realinv, color='blue')
ax.plot(norm_data.realgovt, color='purple')

ax.legend(['realgdp', 'realcons', 'realinv', 'realgovt'])

plt.show()
```

```
fig, axs = plt.subplots(1, 4, figsize=(28, 5))

axs[0].hist(norm_data.realgdp, color='orange')
axs[0].set_title('realgdp')

axs[1].hist(norm_data.realcons, color='blue')
axs[1].set_title('realcons')

axs[2].hist(norm_data.realinv, color='green')
axs[2].set_title('realinv')

axs[3].hist(norm_data.realgovt, color='purple')
axs[3].set_title('realgovt')

plt.show()
```

```
fig, axs = plt.subplots(figsize=(12,8))

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)
axs.hist(norm_data.realgdp, **kwargs)
axs.hist(norm_data.realcons, **kwargs)
axs.hist(norm_data.realinv, **kwargs)
axs.hist(norm_data.realgovt, **kwargs)
axs.legend(['realgdp', 'realcons', 'realinv', 'realgovt'])
```

```
fig, axs = plt.subplots(1, 3, figsize=(24,8))

axs[0].set_xlabel('Real GDP')
axs[0].set_ylabel('Real cons')
axs[0].scatter(norm_data.realgdp, norm_data.realcons)

axs[1].set_xlabel('Real GDP')
axs[1].set_ylabel('Real inv')
axs[1].scatter(norm_data.realgdp, norm_data.realinv)

axs[2].set_xlabel('Real GDP')
axs[2].set_ylabel('Real govt')
axs[2].scatter(norm_data.realgdp, norm_data.realgovt)
```
